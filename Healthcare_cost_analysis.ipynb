```python
import pandas as pd

# Load dataset
df = pd.read_csv("insurance.csv")

# Show first 5 rows
df.head()
```

```python
# Dataset shape
df.shape
```

```python
# Column names
df.columns
```

```python
# Dataset info
df.info()
```

```python
# Average healthcare charges
print("Average Medical Charges:", df['charges'].mean())
```

```python
# Highest medical charge
print("Maximum Medical Charge:", df['charges'].max())
```

```python
# Smoker distribution
df['smoker'].value_counts()
```

```python
smoker_charges = df.groupby('smoker')['charges'].mean()

print(smoker_charges)
```

```python
region_charges = df.groupby('region')['charges'].mean()

print(region_charges)
```

```python
# Average BMI
print("Average BMI:", df['bmi'].mean())
```

```python
import matplotlib.pyplot as plt

df['charges'].plot(kind='hist', bins=30)

plt.title("Healthcare Charges Distribution")
plt.xlabel("Medical Charges")

plt.savefig("charges_distribution.png")

plt.show()
```

```python
smoker_charges.plot(kind='bar')

plt.title("Average Charges: Smokers vs Non-Smokers")
plt.ylabel("Average Medical Charges")

plt.savefig("smoker_cost_comparison.png")

plt.show()
```

```python
region_charges.plot(kind='bar')

plt.title("Average Medical Charges by Region")
plt.ylabel("Average Charges")

plt.savefig("regional_charges.png")

plt.show()
```

```python
plt.scatter(df['age'], df['charges'])

plt.title("Age vs Medical Charges")
plt.xlabel("Age")
plt.ylabel("Charges")

plt.savefig("age_vs_charges.png")

plt.show()
```
